### Integration

```python
def pp(csv_path):
    adata = sc.read_csv(csv_path).T
    sc.pp.filter_genes(adata, min_cells = 10)
    sc.pp.highly_variable_genes(adata, n_top_genes = 2000, subset = True, flavor = 'seurat_v3')
    scvi.model.SCVI.setup_anndata(adata)
    vae = scvi.model.SCVI(adata)
    vae.train()
    solo = scvi.external.SOLO.from_scvi_model(vae)
    solo.train()
    df = solo.predict()
    df['prediction'] = solo.predict(soft = False)
    df.index = df.index.map(lambda x: x[:-2])
    df['dif'] = df.doublet - df.singlet
    doublets = df[(df.prediction == 'doublet') & (df.dif > 1)]
    
    adata = sc.read_csv(csv_path).T
    adata.obs['Sample'] = csv_path.split('_')[2] #'raw_data/GSM5226574_C51ctr_raw_counts.csv'
    
    adata.obs['doublet'] = adata.obs.index.isin(doublets.index)
    adata = adata[~adata.obs.doublet]
    
    
    sc.pp.filter_cells(adata, min_genes=200) #get rid of cells with fewer than 200 genes
    #sc.pp.filter_genes(adata, min_cells=3) #get rid of genes that are found in fewer than 3 cells
    adata.var['mt'] = adata.var_names.str.startswith('mt-')  # annotate the group of mitochondrial genes as 'mt'
    adata.var['ribo'] = adata.var_names.isin(ribo_genes[0].values)
    sc.pp.calculate_qc_metrics(adata, qc_vars=['mt', 'ribo'], percent_top=None, log1p=False, inplace=True)
    upper_lim = np.quantile(adata.obs.n_genes_by_counts.values, .98)
    adata = adata[adata.obs.n_genes_by_counts < upper_lim]
    adata = adata[adata.obs.pct_counts_mt < 20]
    adata = adata[adata.obs.pct_counts_ribo < 2]

    return adata
```

#### Now lets do the integration with scvi

```python
# scvi is not the only tool but we can use it as at now
# Lets save the data as in make a copy, we are just saving the raw data, put it into counts

adata_all.layers['counts'] = adata_all.X.copy()
```

```python
# Now lets normalize the data
# scvi will use the counts data we saved above but other modules will use the raw down here

sc.pp.normalize_total(adata_all,target_sum=1e4)
sc.pp.log1p(adata_all)
adata_all.raw = adata_all
```

```python
adata_all.obs.head()
```

```python
# You would always want the number of genes to be as half as the number of cells
# The genes are not more than our cells in this case
# So we wont further reduce the genes
# Now doing an scvi model for this combined sample

scvi.model.SCVI.setup_anndata(adata_all, layer= "counts",
                             categorical_covariate_keys=['Sample'], #if you had multiple batches, you can add them here 
                             continuous_covariate_keys=['pct_counts_mt', 'total_counts', 'pct_counts_ribo'])
```

```python
model = scvi.model.SCVI(adata_all)
model.train() 
```

```python
#Once the model is trained we want to get a couple of things from the model
# First we can get latent representation

model.get_latent_representation()
```

```python
model.get_latent_representation().shape

# This is the scvi array that represent our data
```

```python
adata_all
```

```python
adata_all.obsm['X_scVI'] = model.get_latent_representation()
```

```python
model.get_normalized_expression(library_size= 1e4)
```

```python
# Saving this as a layer
adata_all.layers['scvi_normalized']= model.get_normalized_expression(library_size= 1e4)
```

```python
# Once you have them saved you can call them back any time
adata_all.layers['scvi_normalized']
```

```python
# CLUSTERING
# Calculate the neighbours

sc.pp.neighbors(adata_all, use_rep= 'X_scVI')
```

```python
# Now that neighbours are calculated we then use the UMAP

sc.tl.umap(adata_all)
sc.tl.leiden(adata_all, resolution=0.5)
```

```python
# Plotting the umap
# One with the cluster label and the other with the sample label so we can see how the integration went

sc.pl.umap(adata_all, color = ['leiden', 'Sample'], frameon = False)
```
